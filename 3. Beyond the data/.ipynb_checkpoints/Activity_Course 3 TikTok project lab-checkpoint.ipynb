#### **Claim status by verification status**

Now, let's create a histogram with four bars: one for each combination of claim status and verification status.

The number of not-verified users is way higher. On the claim status side, the difference is even greater. It makes sense to think that a not-verified user would be more likely to generate a claim. But at the same time, this is proof that being a not-verified user is not directly related to the claim status. Although, it could be a signal that as the platform gathers more not-verified users, there will be a tendency to have more claim videos.

#### **Median view counts by ban status**

```python
# Create a bar plot
ban_status_medians = data.groupby(['author_ban_status']).median(numeric_only=True).reset_index()
plt.figure(figsize=(7,4))
sns.barplot(data=ban_status_medians,
             y='video_view_count',
             x='author_ban_status',
            palette='Set1'
           )
plt.title('Median views by author ban status histogram');
```

Even when active authors are the more common type of user on the platform, they are the type of user that generates the lowest numbers of views. It seems that the content made by banned authors tends to be more engaging.

#### **Total views by claim status**

Let's create a pie graph that depicts the proportions of total views for claim videos and total views for opinion videos.

```python
claim_status_data = data.groupby('claim_status').sum().reset_index()
# Create a pie graph
plt.figure(figsize=(10, 5))
plt.pie(data=claim_status_data, 
        x='video_view_count',
        labels=['Claim videos', 'Opinion videos'], 
        autopct='%1.1f%%', startangle=320
       )
plt.axis('equal')
plt.title('Claim vs opinion views')
# Show the plot
plt.show()
```

Most people are watching claim videos. That means removing claims videos would mean removing the most successful type of video on the platform.

### **Task 4. Determine outliers**

When building predictive models, the presence of outliers can be problematic. For example, if you were trying to predict the view count of a particular video, videos with extremely high view counts might introduce bias to a model. Also, some outliers might indicate problems with how data was captured or recorded.

The analysis performed indicates that a video's engagement level is strongly correlated with its claim status. There's no reason to believe that any of the values in the TikTok data are erroneously captured, and they align with expectation of how social media works: a very small proportion of videos get super high engagement levels. That's the nature of viral content.

```python
for column in ['video_view_count', 'video_like_count', 'video_share_count', 'video_download_count', 'video_comment_count']:
    # IQR calculation
    q1 = data[column].quantile(0.25)
    q3 = data[column].quantile(0.75)
    iqr = q3 - q1
    # Median calculation
    median = data[column].median()
    # Calculate the outlier threshold (median + 1.5 * IQR)
    outlier_threshold = median + 1.5 * iqr
    # Calculate Number of outliers
    number_outliers = data[data[column]>=outlier_threshold][column].count()
    print('Number of outliers, %s: %s' % (column, number_outliers))
```
